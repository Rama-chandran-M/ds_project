```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df = pd.read_csv('rc_merged_01.csv')
df.head(5)
```

```python
import matplotlib.pyplot as plt

# Count how many prospect outcomes appear within each membership renewal decision
ct = pd.crosstab(df['Membership_Renewal_Decision'], df['Prospect_Outcome'])

ax = ct.plot(kind='bar', figsize=(8, 5), width=0.8)
ax.set_title('Prospect Outcome by Membership Renewal Decision')
ax.set_xlabel('Membership_Renewal_Decision')
ax.set_ylabel('Count')
ax.legend(title='Prospect_Outcome')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
