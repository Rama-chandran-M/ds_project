```python
import pandas as pd
df1 = pd.read_csv('rc_final_02.csv')
df2 = pd.read_csv('cc_final_01.csv')
df3 = pd.read_csv('cleaned_emails.csv')
df4 = pd.read_csv('bl_final_01.csv')
```

```python

df1['Prospect_Renewal_Date'] = pd.to_datetime(df1['Prospect_Renewal_Date'], format='mixed')
df1.info()
```
