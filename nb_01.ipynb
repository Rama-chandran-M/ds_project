```python
import pandas as pd
df = pd.read_csv('renewal_calls.csv')
len(df)
```

```python
df = df[df['Analysed_Call'] == 1]
print(len(df))
```

```python
df.drop(columns=['Churn_Category','Complaint_Category','Customer_Reaction_Category','Justification_Category','Reason_For_Renewal_Category','Agent_Response_To_Cancel_Category','Argument_That_Convinced_Customer_to_Stay_Category'], inplace=True)
```

```python
df.info()
```

```python
selected_df = df[["Co_Ref","Call_Date","Membership_Renewal_Decision","Serious_Complaint","Other_Complaint","Discussion_on_Price_Increase","Renewal_Impact_Due_to_Price_Increase",
"Discount_or_Waiver_Requested","Call_Reschedule_Request","Explicit_Competitor_Mention","Explicit_Switching_Intent","Mentioned_Competitors","Desire_To_Cancel","Discount_Offered","Call_Number"]]
selected_df['Call_Date'] = pd.to_datetime(selected_df['Call_Date'],format='%d-%m-%Y')
selected_df.info()
```

```python
# selected_df.to_csv('transformation_01.csv', index=False)
# len(selected_df)
```

```python
selected_df = selected_df.dropna(subset=['Co_Ref'])
len(selected_df)
```

```python
selected_df['Membership_Renewal_Decision'] = selected_df['Membership_Renewal_Decision'].fillna('unknown')
selected_df['Membership_Renewal_Decision'] = selected_df['Membership_Renewal_Decision'].replace('[Yes/No]','unknown')
selected_df['Membership_Renewal_Decision'] = selected_df['Membership_Renewal_Decision'].replace('Yes/No','unknown')
selected_df['Serious_Complaint'] = selected_df['Serious_Complaint'].fillna('unknown')
selected_df['Serious_Complaint'] = selected_df['Serious_Complaint'].replace('[Yes/No]','unknown')
selected_df['Other_Complaint'] = selected_df['Other_Complaint'].fillna('unknown')
selected_df['Other_Complaint'] = selected_df['Other_Complaint'].replace('[Yes/No]','unknown')
selected_df['Other_Complaint'] = selected_df['Other_Complaint'].replace('Not Applicable','unknown')
selected_df['Discussion_on_Price_Increase'] = selected_df['Discussion_on_Price_Increase'].fillna('unknown')
selected_df['Discussion_on_Price_Increase'] = selected_df['Discussion_on_Price_Increase'].replace('**No**','unknown')
selected_df['Discussion_on_Price_Increase'] = selected_df['Discussion_on_Price_Increase'].replace('[Yes/No]','unknown')
selected_df['Renewal_Impact_Due_to_Price_Increase'] = selected_df['Renewal_Impact_Due_to_Price_Increase'].fillna('unknown')
selected_df['Renewal_Impact_Due_to_Price_Increase'] = selected_df['Renewal_Impact_Due_to_Price_Increase'].replace('**No**','No')
selected_df['Renewal_Impact_Due_to_Price_Increase'] = selected_df['Renewal_Impact_Due_to_Price_Increase'].replace('[Yes/No]','unknown')
selected_df['Renewal_Impact_Due_to_Price_Increase'] = selected_df['Renewal_Impact_Due_to_Price_Increase'].replace('Not applicable','unknown')
selected_df['Renewal_Impact_Due_to_Price_Increase'] = selected_df['Renewal_Impact_Due_to_Price_Increase'].str.replace('^Yes.*', 'Yes', regex=True)
```

```python
selected_df['Discount_or_Waiver_Requested'] = selected_df['Discount_or_Waiver_Requested'].fillna('unknown')
selected_df['Discount_or_Waiver_Requested'] = selected_df['Discount_or_Waiver_Requested'].replace('**No**','No')
selected_df['Discount_or_Waiver_Requested'] = selected_df['Discount_or_Waiver_Requested'].replace('[Yes/No]','unknown')
selected_df['Discount_or_Waiver_Requested'] = selected_df['Discount_or_Waiver_Requested'].replace('Not applicable','unknown')
selected_df['Discount_or_Waiver_Requested'] = selected_df['Discount_or_Waiver_Requested'].str.replace('^Yes.*', 'Yes', regex=True)
selected_df['Discount_or_Waiver_Requested'] = selected_df['Discount_or_Waiver_Requested'].str.replace('^No.*', 'No', regex=True)
len(selected_df)
```

```python
selected_df = selected_df[selected_df['Discount_or_Waiver_Requested'].isin(['Yes', 'No', 'unknown'])]
len(selected_df)
```

```python
selected_df['Call_Reschedule_Request'] = selected_df['Call_Reschedule_Request'].fillna('unknown')
selected_df['Call_Reschedule_Request'] = selected_df['Call_Reschedule_Request'].replace('Not Applicable','unknown')
selected_df['Explicit_Competitor_Mention'] = selected_df['Explicit_Competitor_Mention'].fillna('unknown')
selected_df['Explicit_Competitor_Mention'] = selected_df['Explicit_Competitor_Mention'].replace('[No]','No')
len(selected_df)
```

```python
selected_df = selected_df[selected_df['Explicit_Competitor_Mention'].isin(['Yes', 'No', 'unknown'])]
len(selected_df)
```

```python
selected_df['Explicit_Switching_Intent'] = selected_df['Explicit_Switching_Intent'].fillna('unknown')
selected_df['Explicit_Switching_Intent'] = selected_df['Explicit_Switching_Intent'].replace('[No]','No')
selected_df = selected_df[selected_df['Explicit_Switching_Intent'].isin(['Yes', 'No', 'unknown'])]
len(selected_df)
```

```python
selected_df['Mentioned_Competitors'] = selected_df['Mentioned_Competitors'].fillna('unknown')
selected_df['Mentioned_Competitors'] = selected_df['Mentioned_Competitors'].replace('[No]','No')
selected_df = selected_df[selected_df['Mentioned_Competitors'].isin(['Yes', 'No', 'unknown'])]
len(selected_df)
```

```python
selected_df['Desire_To_Cancel'] = selected_df['Desire_To_Cancel'].fillna('unknown')
selected_df['Desire_To_Cancel'] = selected_df['Desire_To_Cancel'].replace('[Renewed / Desired to Cancel]','unknown')
selected_df['Desire_To_Cancel'] = selected_df['Desire_To_Cancel'].str.replace('^Not Discussed.*', 'Not Discussed', regex=True)
selected_df['Desire_To_Cancel'] = selected_df['Desire_To_Cancel'].str.replace('^Not Applicable.*', 'unknown', regex=True)
selected_df['Desire_To_Cancel'] = selected_df['Desire_To_Cancel'].str.replace('^Desired to Cancel.*', 'Desired to Cancel', regex=True)
selected_df['Desire_To_Cancel'] = selected_df['Desire_To_Cancel'].str.replace('^Renew.*', 'Renewed', regex=True)
selected_df['Desire_To_Cancel'] = selected_df['Desire_To_Cancel'].str.replace('^Renewed.*', 'Renewed', regex=True)
selected_df = selected_df[selected_df['Desire_To_Cancel'].isin(['unknown','Not Discussed','Desired to Cancel','Renewed'])]
len(selected_df)
```

```python
selected_df['Discount_Offered'] = selected_df['Discount_Offered'].fillna('unknown')
```

```python
selected_df.to_csv('rc_transformation_01.csv', index=False)
```
